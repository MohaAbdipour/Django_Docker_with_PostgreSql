```python
from google.colab import files

uploaded = files.upload()
audio_file = list(uploaded.keys())[0]
print("Uploaded file:", audio_file)
```

```python
import whisper
import os
import re
from google.colab import files

# Load model
model = whisper.load_model("medium")   # use "base" if you want faster

# Transcribe
result = model.transcribe(
    audio_file,
    language="en",          # change this if needed, or remove it
    word_timestamps=True
)

# Output filename based on uploaded audio name
base_name = os.path.splitext(audio_file)[0]
output_file = f"{base_name}_clean_transcript.txt"

# Format seconds as hh:mm:ss
def fmt_time(sec):
    h = int(sec // 3600)
    m = int((sec % 3600) // 60)
    s = int(sec % 60)
    return f"{h:02d}:{m:02d}:{s:02d}"

# Save one sentence per line with start time
with open(output_file, "w", encoding="utf-8") as f:
    for seg in result["segments"]:
        start_time = fmt_time(seg["start"])
        text = seg["text"].strip()

        # Split into sentences
        sentences = re.split(r'(?<=[.!?])\s+', text)

        for sentence in sentences:
            sentence = sentence.strip()
            if sentence:
                f.write(f"[{start_time}] {sentence}\n")

print(f"Saved as: {output_file}")
files.download(output_file)
```
